```python
import pandas as pd
df = pd.read_csv('realtor_data.csv')
df
```

```python
df2 = df[df.state=='New York']
df3 = df[df.state=='Massachusetts']
df4 = df[df.state=='New Jersey']
df5 = pd.concat([df2, df3, df4], ignore_index=True)
df5
```

```python
df6 = df5[df5.bed<=6]
df7 = df6[df6.bath<=6]
df8 = df7[df7.house_size<=5200]
df9 = df8[df7.house_size>=1782]
df10 = df9[df9.price<=1995000]
df11 = df10[df10.price>=174500]
df11
```

```python
df12 = df11.drop(['status', 'acre_lot', 'zip_code', 'prev_sold_date'], axis='columns')
df12
```

```python
df13 = pd.get_dummies(df12)
df13
```

```python
Y = df13['price']
X = df13.drop(['price'], axis='columns')
X
```

```python
X.drop(['state_Massachusetts', 'state_New Jersey', 'state_New York'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr.score(X_test, y_test)
```
